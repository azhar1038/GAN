```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
igit = load_img("igit_sat.png")
```

```python
igit = img_to_array(igit)
```

```python
igit.shape
```

```python
images=[]
for i in range(igit.shape[0]//256):
    for j in range(igit.shape[1]//256):
        images.append(igit[i*256:(i+1)*256, j*256:(j+1)*256])
```

```python
from numpy import asarray
images = asarray(images, dtype="uint8")
```

```python
images.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))
for i in range(3):
    for j in range(6):
        plt.subplot(3, 6, i*6+j+1)
        plt.axis("off")
        plt.imshow(images[i*6+j])
plt.show()
```
